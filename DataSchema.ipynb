```python
TaskProcessorsData = pd.read_excel(r'C:\Users\guymc\Desktop\CASProject\CAS502Project\production_system_intake_Example01 (4).xlsx',sheet_name='TaskTbl').iloc[:,[1,2,4,5]]
StationData = pd.read_excel(r'C:\Users\guymc\Desktop\CASProject\CAS502Project\production_system_intake_Example01 (4).xlsx',sheet_name='StationTbl').iloc[:,[1,2,7]]
```

```python
TaskProc_df = pd.DataFrame(TaskProcessorsData)
Station_df = pd.DataFrame(StationData)
```

```python
#Create dictionary of ProcessorIDs and StationIDs to cross-walk between the TaskProc_df and the Station_df.
#This dictionary will enable the Sort_Stations function to show alignment between tasks, processors, and stations.

ProcStat_dict = pd.Series(Station_df['StationID'].values, index=Station_df['ProcessorID']).to_dict()
```

```python
def Sort_Stations(df, dictionary, column_name: str, key_column: str):
    df[column_name] = df[key_column].map(dictionary)

    return df
```

```python
Sort_Stations(TaskProc_df, ProcStat_dict, "StationID", "ProcessorID")
```

```python
#Transform the dataframe into a json. The 'records' orientation is best for simulation.
json_string = TaskProc_df.to_json(orient='records')
```
